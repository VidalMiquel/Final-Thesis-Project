```python
import plotly.express as px
import plotly.graph_objects as go
import numpy as np
import pandas as pd
import pickle
```

```python
path = "../Data/Team1/06Stage/Tables/Score/Individual/1_0_individualMetrics.pkl"
```

```python
def load_metrics(file):
    try:
        with open(file, "rb") as f:
            deserializedFile = pickle.load(f)
        return deserializedFile
    except FileNotFoundError:
        print(f"File not found.")
   
```

```python
test = load_metrics(path)
test
```

```python
categories = ["worst","bad","medium","good","excellent"]
cat_values = dict(zip(categories,range(len(categories))))

#tu tens d'un jugador multiples categories per resultats > pots agafar la seva: moda i tens aixo >> df
# també pots fer dos grafics per jugadors segons resultats [0-1,0-2,0-3] i [1-0,2-0,...]
testCut = class_columns.map(lambda x:cat_values[x]) ##en pandas >= 2.1. is map!
testCut
```

```python
fig = go.Figure()

# Iterate over the rows of the class_columns DataFrame using .iloc to avoid KeyError
for i in range(number_of_rows):
    fig.add_trace(
        go.Scatterpolar(
            r=testCut.iloc[i],
            theta=testCut.columns,
            fill='toself',
            opacity=0.5,
            name=f"Player-{testCut.index[i]}",  # Can use the player's name if available
            showlegend=True
        )
    )

fig.update_layout(
    font_size=13,
    legend=dict(
        yanchor="top",
        y=1.,
        xanchor="left",
        x=0.8
    ),
    polar=dict(
        radialaxis=dict(
            angle=45,
            tickangle=45,
            visible=True,
            gridwidth=2,
            range=[0, max(testCut.max(numeric_only=True))],  # Adjust range based on data
            tickvals=list(range(len(testCut.columns))),
            ticktext=categories,
            tickwidth=10
        )
    ),
    title="Player Variables"
)

fig.show()
```
