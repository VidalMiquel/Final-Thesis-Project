```python
def load_metrics(file):
    try:
        with open(file, "rb") as f:
            deserializedFile = pickle.load(f)
        return deserializedFile
    except FileNotFoundError:
        print(f"File not found.")
   
```

```python
test = load_metrics(path)
```

```python
testNo0 = class_columns = test.replace(0, "worst")
```

```python
class_columns = testNo0.xs('Class', level=1, axis=1)
```

```python
number_of_rows = class_columns.shape[0]
```

```python
categories = ["worst","bad","medium","good","excellent"]
cat_values = dict(zip(categories,range(len(categories))))

#tu tens d'un jugador multiples categories per resultats > pots agafar la seva: moda i tens aixo >> df
# també pots fer dos grafics per jugadors segons resultats [0-1,0-2,0-3] i [1-0,2-0,...]
testCut = class_columns.map(lambda x:cat_values[x]) ##en pandas >= 2.1. is map!
```

```python
def plotPolarChart(dataframe, categories):
    """
    Generate a polar chart from a DataFrame.
    
    Args:
    dataframe (pd.DataFrame): The DataFrame containing the data.
    categories (list): The list of categories for the polar axis.
    
    Returns:
    go.Figure: The Plotly Figure object.
    """
    fig = go.Figure()

    # Iterate over the rows of the DataFrame
    for i in range(len(dataframe)):
        fig.add_trace(
            go.Scatterpolar(
                r=dataframe.iloc[i],
                theta=dataframe.columns,
                fill='toself',
                opacity=0.5,
                name=f"Player-{dataframe.index[i]}",  # Use player's name if available
                showlegend=True
            )
        )

    fig.update_layout(
        font_size=13,
        legend=dict(
            yanchor="top",
            y=1.,
            xanchor="left",
            x=0.8
        ),
        polar=dict(
            radialaxis=dict(
                angle=45,
                tickangle=45,
                visible=True,
                gridwidth=2,
                range=[0, max(dataframe.max(numeric_only=True))],  # Adjust range based on data
                tickvals=list(range(len(dataframe.columns))),
                ticktext=categories,
                tickwidth=10
            )
        ),
        title="Player Variables"
    )

    return fig
```

```python
fig = plotPolarChart(testCut, categories)
fig.show()
```
