```python
import pandas as pd
from tabulate import tabulate
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#Define path to the experiments
Barcelona = "../Data/Team1/03Stage/finalMetadataBarcelona.csv"
RayoVallecano = "../Data/Team2/03Stage/finalMetadataRayo Vallecano.csv"
```

```python
try:
    # Read the CSV files into DataFrames
    metadataBarcelona = pd.read_csv(Barcelona, sep=",")
    metadataRayoVallecano = pd.read_csv(RayoVallecano, sep=",")
    # Now metadataManUnited and metadataManCity contain the data from the CSV files
except FileNotFoundError:
    print("One of the files does not exist or the path is incorrect.")
except Exception as e:
    print("An error occurred:", str(e))
```

```python
def addDifferenceToTable(existingTable, differenceValues):
    numDiff = len(differenceValues)
    # Create a DataFrame with the difference values and initial win, lose, draw values set to 0
    diffDF = pd.DataFrame({'Score': differenceValues, 'win': [0]*numDiff, 'lose': [0]*numDiff, 'draw': [0]*numDiff})
    # Concatenate the existing table with the difference DataFrame
    resultTable = pd.concat([existingTable, diffDF], ignore_index=True)
    # Sort the resulting table by the Score column
    resultTableSorted = resultTable.sort_values(by="Score")
    return resultTableSorted
```

```python
def getDifference(dataFrame, table):
    # Extract unique score values from metadataRayoVallecano
    allUniqueResults = dataFrame["Score"].dropna().unique()
    noNFResults = allUniqueResults[allUniqueResults != "NF"]

    # Extract scores from the table
    scores = table["Score"]

    # Calculate the difference
    difference = list(set(noNFResults) - set(scores))

    return difference
```

```python
def generateTable(dataFrame):
    nfIndices = dataFrame.index[dataFrame["Score"] == "NF"]
    # Filter rows with non-NF indices
    filteredRows = dataFrame[dataFrame.index.isin(nfIndices - 1)].copy()  # Create a copy to avoid warnings

    # Define a function to determine the outcome
    def determineOutcome(score):
        homeScore, awayScore = map(int, score.split("_"))
        difference = homeScore - awayScore
        if difference > 0:
            return "win"
        elif difference < 0:
            return "lose"
        else:
            return "draw"

    # Apply the determineOutcome function to create the "outcome" column
    filteredRows["outcome"] = filteredRows["Score"].apply(determineOutcome)

    # Count the occurrences of each outcome for each `idFiles`
    outcomeCounts = filteredRows.groupby(["Score", "outcome"]).size().unstack(fill_value=0).reset_index()

    # Generate the desired table
    table = pd.DataFrame({
        "Score": outcomeCounts["Score"],
        "win": outcomeCounts["win"],
        "lose": outcomeCounts["lose"],
        "draw": outcomeCounts["draw"]
    })

    return table

    return table
```

FCBarcelona

```python
table = generateTable(metadataBarcelona)
difference = getDifference(metadataBarcelona, table)
finalTableBarcelona = addDifferenceToTable(table,difference)
```
