```python
def loadPlayersList(playersList):
    try:
        with open(playersList, 'rb') as file:
            # Load data from the file
            playersListData = pickle.load(file)
        return playersListData
    except FileNotFoundError:
        print("File not found:", playersList)
        return None
    except Exception as e:
        print("An error occurred while loading the players list:", str(e))
        return None
```

```python
def loadPlayersDataFrame(playersDict):
    try:
        players_df = pd.DataFrame(list(playersDict.items()), columns=['Player ID', 'Player Name'])
        players_df.reset_index(drop=True, inplace=True)
        return players_df
    except Exception as e:
        print("An error occurred while loading the players DataFrame:", str(e))
        return None
```
